## Data preprocessing
All stop words were removed from the documents

```python
# loading the documents
# taking the pdfs from the folder and extracting all the documents within them
def docs_loader():
    docs = []
    count = 0
    for filename in os.listdir(path):
        if filename.endswith('.pdf'):
            count = count + 1
            print(f"{filename} {count}")
            file_path = os.path.join(path, filename)
            loader = PyPDFLoader(file_path)
            docs.extend(loader.load())
    print(f"Total documents loaded: {len(docs)}")
    return docs
```

```python
docs = docs_loader()
```

```python
def splitting_text(docs):
    text_splitter = RecursiveCharacterTextSplitter(chunk_size=1000,
                                                   chunk_overlap=100,
                                                   length_function=len,
                                                   add_start_index=True
                                                   )

    # Pre-processing: removes all stop words from the document before embedding it
    # numbers are kept because graphs will be drawn from said data
    chunks = text_splitter.split_documents(docs)
    # preprocessing removing stop words from each chunk
    cleaned = []
    for chunk in chunks:
        # tokenizing the text
        words = word_tokenize(chunk.page_content.lower())
        # remove the stop words
        filtered_words = [word for word in words if word not in stop_words]

        cleaned_text = ' '.join(filtered_words)

        cleaned_chunk = Document(page_content=cleaned_text, metadata=chunk.metadata)
        cleaned.append(cleaned_chunk)

    return cleaned
```

```python
chunks = splitting_text(docs)
```

## Exploratory data analysis
Use of graphs such as word cloud and histograms

```python
# Pre-processing: removes stop words, punctuations, numbers and other irrelevant data when drawing the graphs
def clean_text(text):
    # Remove punctuation
    text = text.translate(str.maketrans('', '', string.punctuation))
    # Remove numbers
    text = ''.join([i for i in text if not i.isdigit()])
    # Convert to lowercase and split into words
    words = text.lower().split()
    # Remove stop words and 'et al'
    words = [word for word in words if word not in stop_words and word != 'et']
    return ' '.join(words)
```

## Document Embedding and creation of chroma db

```python
def create_or_load_database(chunks=None):
    collection = chroma_client.get_or_create_collection("pesticides_docs")

    if chunks and collection.count() == 0:
        print("Adding documents to the collection...")
        embeddings = OpenAIEmbeddings()

        # Batch size
        batch_size = 500

        for i in range(0, len(chunks), batch_size):
            batch = chunks[i:i + batch_size]

            ids = [f"doc_{j}" for j in range(i, min(i + batch_size, len(chunks)))]
            texts = [chunk.page_content for chunk in batch]
            metadatas = [chunk.metadata for chunk in batch]

            embeddings_list = embeddings.embed_documents(texts)

            collection.add(
                ids=ids,
                embeddings=embeddings_list,  # Use combined features instead of just embeddings
                metadatas=metadatas,
                documents=texts
            )
            print(f"Added batch {i // batch_size + 1} to the collection.")

        print(f"Added {len(chunks)} documents to the collection in total.")

    return collection
```

```python
def check_chroma_collection():
    try:
        collection = chroma_client.get_or_create_collection("pesticides_docs")
        count = collection.count()
        print(count)
        if count > 0:
            print(f"Chroma collection found with {count} documents")
            return collection
        else:
            print("Chroma collection exists but is empty. Adding documents...")
            # docs = docs_loader()
            # chunks = splitting_text(docs)
            # perform_eda(chunks)
            collection = create_or_load_database(chunks=chunks)
            print(f"Added {collection.count()} documents to the collection")
            return collection
    except Exception as e:
        print(e)
        return None
```

## Training of models

```python
def setup_retrieval():
    embeddings = OpenAIEmbeddings()
    db = Chroma(client=chroma_client, collection_name='pesticides_docs', embedding_function=embeddings)

    retriever = db.as_retriever(search_kwargs={"k": 3})

    # specifying of accurate prompts to be used, by the open AI model
    prompt_template = """Use the following pieces of context to answer the question at the end.
    Provide detailed information about pesticide application procedures, safety considerations, and environmental impact.
    Include specific details about the pesticide if available, such as its chemical composition and recommended usage.

    Consider the following aspects in your answer:
    1. Application method and timing
    2. Safety precautions for farmers
    3. Environmental impact (soil, water, air)
    4. Effects on beneficial insects and wildlife
    5. Pre-harvest interval and chemical residue concerns
    6. Alternatives or integrated pest management strategies

    If the question is about a specific pesticide and you don't have information about it, suggest similar pesticides or general best practices.
    
    If you don't know the answer, say that you don't know, don't try to make up an answer.
    
    If the user greets you, response back with a greeting

    {context}

    Question: {question}

    Answer: """

    PROMPT = PromptTemplate(
        template=prompt_template, input_variables=["context", "question"]
    )
    chain_type_kwargs = {"prompt": PROMPT}
    return RetrievalQA.from_chain_type(
        llm=ChatOpenAI(),
        chain_type="stuff",
        retriever=retriever,
        chain_type_kwargs=chain_type_kwargs,
        return_source_documents=True
    )
```

```python
qa = setup_retrieval()
```

## Evaluation and Optimisation of model


```python
# Evaluation: calculates the models response using the rogue score
def calculate_rogue(generated_answer, expected_answer):
    print('Generating rogue score')
    # ROUGE Score
    rouge = Rouge()
    rouge_scores = rouge.get_scores(generated_answer, expected_answer)[0]

    return {
        "rouge-1": rouge_scores['rouge-1']['f'],
        "rouge-2": rouge_scores['rouge-2']['f'],
        "rouge-l": rouge_scores['rouge-l']['f'],
    }
```

```python
from nltk.translate.bleu_score import sentence_bleu
from nltk.tokenize import word_tokenize
from nltk.translate.bleu_score import sentence_bleu, SmoothingFunction

def calculate_bleu_score(reference, candidate):
    reference_tokens = reference.split()
    candidate_tokens = candidate.split()
    
    # Option 1: Use lower n-gram orders
    weights = (0.25, 0.25, 0.25, 0.25)  # Equal weights for 1-gram to 4-gram
    
    # Option 2: Use smoothing function
    smoothie = SmoothingFunction().method1  # You can try different methods (method0 to method7)
    
    return sentence_bleu([reference_tokens], candidate_tokens, weights=weights, smoothing_function=smoothie)
```

```python
# Evaluation passing and testing of expected and generated answer
def evaluation_of_response(qa):
    # Predefined set of questions and expected answers for evaluation
    evaluation_set = [
            {
                "question": "What are the main safety precautions when applying pesticides?",
                "expected_answer": "The main safety precautions when applying pesticides include wearing proper "
                                   "protective equipment such as gloves, goggles, and respirators; following label "
                                   "instructions carefully; avoiding application in windy conditions; washing hands and "
                                   "clothing after use; and storing pesticides in a secure, well-ventilated area away "
                                   "from food and water sources."
            },
            {
                "question": "How do pesticides impact beneficial insects?",
                "expected_answer": "Pesticides can negatively impact beneficial insects by directly killing them, "
                                   "reducing their food sources, or contaminating their habitats. This can disrupt local "
                                   "ecosystems, affecting pollination processes and natural pest control. Broad-spectrum "
                                   "insecticides are particularly harmful, often killing both target pests and beneficial "
                                   "insects indiscriminately."
            },
            {
                "question": "What is integrated pest management (IPM) and how does it relate to pesticide use?",
                "expected_answer": "Integrated Pest Management (IPM) is an ecosystem-based approach to crop protection "
                                   "that combines different management strategies and practices to grow healthy crops and "
                                   "minimize the use of pesticides. It involves monitoring pest levels, using biological "
                                   "controls, implementing cultural practices, and using pesticides only when necessary. "
                                   "IPM aims to reduce pesticide use, minimize environmental impact, and maintain "
                                   "economic viability of crop production."
            },
            {
                "question": "What are the potential environmental impacts of pesticide runoff?",
                "expected_answer": "Pesticide runoff can have severe environmental impacts, including water pollution in "
                                   "rivers, lakes, and groundwater. This can lead to the death of aquatic organisms, "
                                   "disruption of aquatic ecosystems, and bioaccumulation in the food chain. It may also "
                                   "contaminate drinking water sources, affect soil microorganisms, and contribute to the "
                                   "development of pesticide-resistant species."
            },
        ]
    
    total_bleu_score = 0

    for item in evaluation_set:
        query = item["question"]
        expected_answer = item["expected_answer"]

        try:
            response = qa({"query": query})
            answer = response['result']

            print("\nQuestion:", query)
            print("Generated Answer:", answer)
            print("Expected Answer:", expected_answer)

            rogue_result = calculate_rogue(answer, expected_answer)
            bleu_score = calculate_bleu_score(expected_answer, answer)
            print(f"BLEU Score: {bleu_score}")
            print(rogue_result)

            total_bleu_score += bleu_score

        except Exception as e:
            print(f"An error occurred: {e}")
            
    average_bleu_score = total_bleu_score / len(evaluation_set)
    print(f"\nAverage BLEU Score: {average_bleu_score}")
```

```python
evaluation_of_response(qa)
```
